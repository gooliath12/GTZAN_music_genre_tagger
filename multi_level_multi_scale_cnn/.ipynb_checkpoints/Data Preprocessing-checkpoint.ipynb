# Data Preprocessing

This Script is used for convert each song clip in GTZAN dataset to a (mel-)spectrogram. 

After **downloading the dataset and specifying the path to dataset**, run all cells below to convert data split it to train and test sets. The train/test split is recorded in `gtzan_train.csv` and `gtzan_test.csv`. Of course you can implement a new train/test split.

```python
import os
import numpy as np
import pandas as pd
import librosa
import time
import random
from sklearn.model_selection import train_test_split
```

```python
# PATH TO ORIGIN DATASET, MODIFY THIS TO THE PATH OF DATASET ON YOUR COMPUTER!
PATH_DATASET = '/Users/pengguo/Desktop/coms4995/Project/datasets/gtzan_genre/genres/'
# Path to root directory
PATH_ROOT = './'

# Path to two CSV files, both of which are already included in the repository
TRAIN_CSV = PATH_ROOT + 'gtzan_train.csv'
TEST_CSV = PATH_ROOT + 'gtzan_test.csv'

# PARAMETERS
SR = 12000 # sample rate: 12KHz
len_src = 29. # duration: 29 secs
ref_n_src = 12000 * 29
```

## Convert Data to Mel-spectrograms

The code below are used for convert each song clips to mel-spectrograms. 

For the convience to reproduce our result, we have recorded our train/test data split in `gtzan_train.csv` and `gtzan_test.csv`. Each spectrogram would be stored in different .npy files according to these CSV files.

```python
import librosa
```

```python
# Build Data Paths for storing spectrograms
if not os.path.exists(PATH_ROOT + 'dataset/'):
    os.mkdir(PATH_ROOT + 'dataset/')

X_TRAIN = PATH_ROOT + 'dataset/X_train.npy'
Y_TRAIN = PATH_ROOT + 'dataset/Y_train.npy'
X_TEST = PATH_ROOT + 'dataset/X_test.npy'
Y_TEST = PATH_ROOT + 'dataset/Y_test.npy'
```

```python
### Helper Functions ###
def gen_filepaths(df):
    '''
    A Generator reading file paths from dataframe.
    Inputs:
        - df: Dataframe converted from .csv file.
    Return:
        - paths to each audio file
    '''
    for filepath in df['filepath']:
        yield os.path.join(PATH_DATASET, filepath)


def _load_audio(path):
    '''
    Load audio file at path with sampling rate=SR, duration=len_src, and return it
    '''
    src, sr = librosa.load(path, sr=SR, duration=len_src * SR / float(SR))
    
    # Unify the length of src (in case its longer than ref_n_src)
    src = src[:ref_n_src]
    result = np.zeros(ref_n_src)
    result[:len(src)] = src[:ref_n_src]
    
    # transform src to mel-spectrogram
    src_ms = librosa.feature.melspectrogram(y=result, n_fft=512, hop_length=256, power=2.0, sr=SR, n_mels=96, \
                                   fmin=0.0, fmax=6000)
    
    return src_ms


def gen_X(PATH_CSV, PATH_SAVE, NUM_DATA):
    '''
    Convert GTZAN audio files to a single .npy file.
    shape of X: (1000, 1, 348000)
    '''
    print 'Start generating X...'
    start_time = time.time()
    df = pd.DataFrame.from_csv(PATH_CSV) # load csv
    filepaths = gen_filepaths(df)  # filepaths: generator
    
    raw_data = np.array([_load_audio(filepaths.next()) for _ in xrange(NUM_DATA)])
    data = raw_data[:, np.newaxis, :, :] # "Insert" one dimension
    np.save(PATH_SAVE, data)
    print 'DONE! in {:6.4f} sec'.format(time.time() - start_time)
    return data


def gen_Y(PATH_CSV, PATH_SAVE):
    '''
    Generate labels for GTZAN dataset.
    '''
    df = pd.DataFrame.from_csv(PATH_CSV) # load csv
    data = np.array(df['label'])
    np.save(PATH_SAVE, data)
    return data  
```

```python
# Generate X_train, Y_train
X_train = gen_X(TRAIN_CSV, X_TRAIN, 900)
Y_train = gen_Y(TRAIN_CSV, Y_TRAIN)

# Generate X_test, Y_test
X_test = gen_X(TEST_CSV, X_TEST, 100)
Y_test = gen_Y(TEST_CSV, Y_TEST)
```

## Split Training & Test Data

Split training and test data to segments in different length for local feature extractor (CNNs) training. The segment sizes are 20, 30, 60, 120, 240.

```python
def chop_data(seg_length, X, Y):
    '''
    Chop original mel-spectrograms to segments.
    '''
    num_pics, num_channels, x, y = X.shape
    # num_pics: Number of pictures
    # num_channels: number of channels
    # x: number of mel-bins (vertical)
    # y: number of time periods (horizontal)
    
    X_new = np.zeros((num_pics * (y / seg_length), num_channels, x, seg_length))
    idx = 0
    for i in xrange(num_pics):
        for j in xrange(y / seg_length):
            X_new[idx, :, :, :] = X[i, :, :, j*seg_length:(j+1)*seg_length]
            idx += 1       
    
    # Check if chopped correctly
    assert(np.all(X_new[-1, -1, :, :] == X[-1, -1, :, (y/seg_length - 1)*seg_length: (y/seg_length) * seg_length]))
    
    # Check Y's shape
    assert(Y.shape == (num_pics,))
    
    Y_new = np.zeros(num_pics * (y / seg_length))
    idx = 0
    for i in xrange(num_pics):
        for _ in xrange(y / seg_length):
            Y_new[idx] = Y[i]
            idx += 1
    
    # Store data segments
    if num_pics == 100:
        np.save(PATH_ROOT + 'dataset/X_test_seg{}.npy'.format(seg_length), X_new)
        np.save(PATH_ROOT + 'dataset/Y_test_seg{}.npy'.format(seg_length), Y_new)
    if num_pics == 900:
        np.save(PATH_ROOT + 'dataset/X_train_seg{}.npy'.format(seg_length), X_new)
        np.save(PATH_ROOT + 'dataset/Y_train_seg{}.npy'.format(seg_length), Y_new)
    
    return 
```

```python
X, Y = np.load(PATH_ROOT + 'dataset/X_test.npy'), np.load(PATH_ROOT + 'dataset/Y_test.npy')
for seg_length in [20, 30, 60, 120, 240]:
    chop_data(seg_length, X, Y)
```

# Spectrogram Visualization

The code below are used for visualize spectrograms we've just got above.

```python
import librosa
from librosa import display
import matplotlib.pyplot as plt

# Path to store the figures, remember to change to path on your computer
PATH_FIGURE = '/Users/pengguo/Desktop/coms4995/Project/figures/'

# Load sound file
# Also, remember to change to your paths
path_jazz = "/Users/pengguo/Desktop/coms4995/Project/datasets/gtzan_genre/genres/jazz/jazz.00033.au"
path_classical = "/Users/pengguo/Desktop/coms4995/Project/datasets/gtzan_genre/genres/classical/classical.00079.au"
```

```python
def show_spectrogram(path):
    '''
    Visualize the spectrogram that `path` is pointed to.
    '''
    y, sr = librosa.load(path, sr=SR, duration=len_src * SR / float(SR))

    # Let's make and display a mel-scaled power (energy-squared) spectrogram
    S =  librosa.feature.melspectrogram(y=y, n_fft=512, hop_length=256, power=2.0, sr=SR, n_mels=96, \
                                       fmin=0.0, fmax=6000)

    # Convert to log scale (dB). We'll use the peak power as reference.
    log_S = librosa.logamplitude(S, ref_power=np.max)

    # Make a new figure
    fig = plt.figure(figsize=(12,4))

    # Display the spectrogram on a mel scale
    # sample rate and hop length parameters are used to render the time axis
    librosa.display.specshow(log_S, sr=SR, x_axis='time', y_axis='mel')

    # Put a descriptive title on the plot
    plt.title('mel power spectrogram')

    # draw a color bar
    plt.colorbar(format='%+02.0f dB')

    # Make the figure layout compact
    plt.tight_layout()
    
    # Display image
    plt.show()
    
    # Save image
    if path.split('/')[-2] == "jazz":
        fig.savefig(PATH_FIGURE + 'jazz_full.png')
    elif path.split('/')[-2] == "classical":
        fig.savefig(PATH_FIGURE + 'classical_full.png')
    
    return
```

```python
show_spectrogram(path_jazz)
show_spectrogram(path_classical)
```
